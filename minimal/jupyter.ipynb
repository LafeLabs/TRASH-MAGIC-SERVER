```python
# total recursion
import urllib.request
import urllib.parse
import matplotlib.pyplot as plt
import numpy as np
import time
import json

def pingurl(url_value):
    try:
        with urllib.request.urlopen(url_value, timeout=5) as response:
            # .read() gets raw bytes, .decode() turns it into text
            html_content = response.read().decode('utf-8')
            return html_content
    except urllib.error.URLError as e:
        print(f"Failed to fetch URL. Reason: {e.reason}")
        
```

```python
timestamp = str(round(time.time()))
data_packet = {}
data_packet['message'] = "I AM THE FUNGUS"
data_packet['origin'] = "DIRT WIZARD LAPTOP"
data_packet['timestamp'] = timestamp
data_packet['baseurl'] = 'http://localhost/TRASH-MAGIC-SERVER/minimal/'
data_packet['filename'] = 'data/data-packet-' + timestamp + '.json'
data_packet_string =  urllib.parse.quote_plus(json.dumps(data_packet))
pingurl(data_packet['baseurl'] + 'save-file-get.php?data=' + data_packet_string + '&filename=' + data_packet['filename'])
```
